TEST AND TRAINING DATA DEFINITION

```python
training_data = data.sample(frac=0.8, random_state=25)
testing_data = data.drop(training_data.index)
```

```python
X_train= training_data.drop(['label'], axis =1)
X_test= testing_data.drop(['label'], axis= 1)
y_train= training_data['label'].copy()
y_test= testing_data['label'].copy()
```

DECISION TREE- Model 1

```python
from sklearn.tree import DecisionTreeClassifier #Desision 
clf = DecisionTreeClassifier(random_state=2)
clf.fit(X_train,y_train)
```

```python
y_test_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix_a = confusion_matrix(y_test, y_test_pred, labels=[0,1])
print(confusion_matrix_a)
```

```python
import sklearn
from sklearn.metrics import accuracy_score
sklearn.metrics.accuracy_score(y_test, y_test_pred)*100
```

```python
specificity_a= confusion_matrix_a[0,0]/(confusion_matrix_a[0,0]+confusion_matrix_a[0,1])
print(specificity_a)
```

```python
sensitivity_a = confusion_matrix_a[1,1]/(confusion_matrix_a[1,0]+confusion_matrix_a[1,1])
print(sensitivity_a)
```

LOGISTIC REGRESSION - Model 2

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(class_weight='balanced')
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
sklearn.metrics.accuracy_score(y_test, predictions)*100
```
